```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import pickle
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Лабы нейронки/data.xlsx")

df = df[["ХСН", "САД", "КСР", "QRS", "HCT", "PLT", "СКФ"]]
# Delete the lines without data
df = df.dropna()
df = df.to_numpy()

X = df[:, 1:]
y = df[:, 0]

print(df[0])
```

```python
# Devide data to test and training data
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.32, random_state=42)

# Model training
model = LogisticRegression()
model.fit(X_train, y_train)

# Prediction
y_pred = model.predict(X_test)

# Оценка модели
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
# Save the prediction in file
predictions_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
predictions_df.to_csv('predictions.csv', index=False)
```

```python
# Predict the input iris
random_iris = list(map(float, input().split()))
p = model.predict_proba([random_iris])
iris_p = p[0][1]

print(f'Random patient features: {random_iris}')
print(f'Probability of being XCH: {iris_p}')
```

```python
# Сохранение модели
with open('drive/MyDrive/Colab Notebooks/logistic_regression_model.pkl', 'wb') as file:
    pickle.dump(model, file)
```
