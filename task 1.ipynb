```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets
```

```python
df = pd.read_excel("drive/MyDrive/Colab Notebooks/data.xlsx")

df = df[["ХСН", "САД", "КСР", "QRS", "HCT", "PLT", "СКФ"]]
# 0 - больной, 1 - здоровый
# Удаляем строки без значений
df = df.dropna()

result = df.columns[0]
params = df.columns[1:]

df.head()
```

```python
sns.pairplot(df, hue="ХСН")
```

```python
def switch_plot(plot_type):
  sns.boxplot(df, x=result, y=plot_type, hue=result)
  plt.show()


# Создание виджета для переключения между графиками
plot_type_widget = widgets.Dropdown(
    options=params,
    value=params[0],
    description='Plot Type:',
    disabled=False,
)

# Использование interact для создания интерактивного переключения
interact(switch_plot, plot_type=plot_type_widget)
```

```python
def switch_plot(plot_type):
  sns.stripplot(df, x=result, y=plot_type, hue=result)
  plt.show()


# Создание виджета для переключения между графиками
plot_type_widget = widgets.Dropdown(
    options=params,
    value=params[0],
    description='Plot Type:',
    disabled=False,
)

# Использование interact для создания интерактивного переключения
interact(switch_plot, plot_type=plot_type_widget)
```

```python
def switch_plot(plot_type):
  sns.violinplot(df, x=result, y=plot_type, hue=result)
  plt.show()


# Создание виджета для переключения между графиками
plot_type_widget = widgets.Dropdown(
    options=params,
    value=params[0],
    description='Plot Type:',
    disabled=False,
)

# Использование interact для создания интерактивного переключения
interact(switch_plot, plot_type=plot_type_widget)
```

```python
def switch_plot(plot_type):
  sns.swarmplot(df, x=result, y=plot_type, hue=result, palette="dark:red")
  sns.violinplot(df, x=result, y=plot_type, hue=result)
  plt.show()


# Создание виджета для переключения между графиками
plot_type_widget = widgets.Dropdown(
    options=params,
    value=params[0],
    description='Plot Type:',
    disabled=False,
)

# Использование interact для создания интерактивного переключения
interact(switch_plot, plot_type=plot_type_widget)
```
